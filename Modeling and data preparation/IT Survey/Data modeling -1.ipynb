####**6. Faça 5 análises multivariadas e apresente sua conclusão. Dica: utilize a função "pivot table" do Pandas; as análises podem ser entre variáveis categóricas, variáveis numéricas ou categórica e numérica.**

####1. Quantidade de pessoas nas cidades por gênero

```python
pd.pivot_table(df_full, values='Timestamp', index='City', columns='Gender', aggfunc='count', fill_value=0)
```

####Análise: Podemos analisar que a quantidade de pessoas de gênero masculino e feminino nas cidades mostrou que, em geral, há uma diferença significativa na quantidade de homens e mulheres.

####2. Media de idade por senioridade

```python
df_full['Seniority level'] = df_full['Seniority level'].replace(to_replace= ['student','Entry level','Working Student','intern','Student','Intern','C-Level'],value='Junior')
df_full['Seniority level'] = df_full['Seniority level'].replace(to_replace= ['No level ', 'Self employed','no idea, there are no ranges in the firm '],value='No level')
df_full['Seniority level'] = df_full['Seniority level'].replace(to_replace= ['Principal','Director','VP','CTO','Key'],value='Head')
df_full['Seniority level'] = df_full['Seniority level'].replace(to_replace= ['Manager','Work Center Manager','C-level executive manager'],value='Lead')
df_full['Seniority level'].value_counts()
```

```python
pd.pivot_table(df_full, values='Age', index='Seniority level', aggfunc=np.mean)
```

####Análise: A análise sobre a média de idade por senioridade mostrou que, a senioridade de uma pessoa aumenta com a idade. No entanto, a pesquisa também mostrou que a média de idade por senioridade varia de acordo com o nível de senioridade (no caso do "No level").

####3. Duração de contrato por gênero

```python
pd.pivot_table(df_full, columns='Сontract duration', index='Gender', aggfunc='size', fill_value=0)
```

####Análise: A análise sobre a duração de contratos por gênero mostrou que, em geral, os contratos entre homens e mulheres têm durações diferentes de forma significativa.

####4. Média salarial pelo tamanho da companhia

```python
pd.pivot_table(df_full, values='Yearly brutto salary (without bonus and stocks) in EUR', index='Company size', aggfunc=np.mean, fill_value=0)
```

####Análise: A análise indica que existe uma correlação entre o tamanho da companhia e a média salarial dos funcionários. Em geral, as empresas de maior porte tendem a ter uma média salarial mais alto do que as empresas menores.

####***Faça um gráfico contendo a quantidade de entrevistas realizadas por mês para cada um dos 3 anos de pesquisa (exclua as linhas onde o ano da pesquisa é diferente do ano do timestamp).***

```python
df_18['Month'] = pd.to_datetime(df_18['Timestamp']).dt.month
df_19['Month'] = pd.to_datetime(df_19['Timestamp']).dt.month
df_20['Month'] = pd.to_datetime(df_20['Timestamp']).dt.month
```

```python
base18_count = df_18.groupby('Month').size()
base19_count = df_19.groupby('Month').size()
base20_count = df_20.groupby('Month').size()
```

```python
data = {
    'Base 2018': base18_count,
    'Base 2019': base19_count,
    'Base 2020': base20_count
}
df = pd.DataFrame(data)

# Plotando o gráfico de barras
ax = df.plot(kind='bar', figsize=(10, 6), width=0.8)

# Definindo o título e os rótulos
ax.set_title('Quantidade de entrevistas realizadas por mês')
ax.set_xlabel('Mês')
ax.set_ylabel('Número de entrevistas')

# Exibindo o gráfico
plt.show()
```
